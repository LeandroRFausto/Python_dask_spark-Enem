```python
# Importa as bibliotecas necessárias
import pandas as pd
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Cria um diretório para armazenar o conteúdo do ENEM 2019
base_path = "./enem2019"
os.makedirs(base_path, exist_ok=True)

# Define a url e faz o download dos dados
url = "http://download.inep.gov.br/microdados/microdados_enem_2019.zip"
filebytes = BytesIO(
    requests.get(url, stream=True).content
)

# Extrai o conteúdo do zipfile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall(base_path)
```

# Ler e processar os dados do ENEM 2019


```python
file_path = os.path.join("enem2019", "DADOS", "MICRODADOS_ENEM_2019.csv")
file_path
```

```python
import dask.dataframe as ddf
```

```python
enem = ddf.read_csv(
    file_path,
    sep=";", encoding="latin1",
    dtype={'CO_MUNICIPIO_NASCIMENTO': 'float64',
       'CO_UF_NASCIMENTO': 'float64',
       'NO_MUNICIPIO_ESC': 'object',
       'SG_UF_ESC': 'object',
       'NU_IDADE': 'float64'}
)
```

enem

```python
enem.head(10)
```

## Filtra os alunos que residam no Estado do RJ

```python
enemrj = enem.loc[enem.SG_UF_RESIDENCIA == "RJ"]
```

```python
enemrj
```

## Calcula a media de idade e a média da nota em matemática por sexo

```python
consulta = enemrj.groupby("TP_SEXO").agg({
    "NU_IDADE": "mean",
    "NU_NOTA_MT": "mean"
})
```

```python
consulta
```

```python
consulta.compute()
```

```python
consulta2 = enemrj.groupby(["TP_SEXO", "NO_MUNICIPIO_RESIDENCIA"]).agg({
    "NU_IDADE": "mean",
    "NU_NOTA_MT": "mean"
})
```

```python
consulta2.compute()
```
